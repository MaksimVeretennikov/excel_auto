```python
import pandas as pd
import xlwt
import math
import numpy as np
#path_name = "/Users/maksimveretennikov/Downloads/Svodnaya_tablitsa_NGS.xlsx"
path_name = "/Users/maksimveretennikov/Downloads/ЯрБройлерИнкубин-2.xlsm"
#sheet_name = "Таблица 2 опыт разделенная"
sheet_name = "Лист2"
file = pd.read_excel(path_name, sheet_name)
index_good = []
good_list = ["Bacteroidetes", "Ruminococcaceae", "Lachnospiraceae", "Eubacteriaceae", "Peptostreptococcaceae",
            "Clostridiaceae", "Thermoanaerobacterales", "Сумма целлюлозолитиков", "Veillonellaceae", 
            "Сумма бацилл", "Bifidobacteriales", "Lactobacillales", "Actinomycetales", "Enterobacteriaceae"]
good_str = ["Bacteroidetes", "Thermoanaerobacterales", "Bifidobacteriales", "Lactobacillales", "Actinomycetales",
            "Ruminococcaceae", "Lachnospiraceae", "Eubacteriaceae", "Peptostreptococcaceae", "Veillonellaceae", 
            "Enterobacteriaceae"]
search_clostrid = "Clostrid"
search_bacil = "Bacilla"
search_paen = "Paenibacillaceae"
good_phylum = good_str[0]
good_orders = good_str[1:5]
good_families = good_str[5:]
for index, row in file.iterrows():   
    if ((type(row[1]) != float and row[1] in good_phylum)
        or (type(row[3]) != float and row[3] in good_orders) 
        or (type(row[4]) != float and row[4] in good_families)
        or (type(row[4]) != float and row[4].find(search_clostrid) != -1)
        or (type(row[4]) != float and row[4].find(search_bacil) != -1)
        or (type(row[4]) != float and row[4].find(search_paen) != -1)):
        index_good.append(index)
good_bacteries = file.loc[index_good, :]
#good_bacteries = good_bacteries.drop(good_bacteries.columns[[13, 14, 15, 16]], axis=1)
#print(good_bacteries)

workbook = xlwt.Workbook()
gb = workbook.add_sheet('Good bactries')
font = xlwt.Font() # Create the Font
font.name = 'Times New Roman'
font.height = 220 # 11 * 20, for 11 point
style = xlwt.XFStyle() # Create the Style
style.font = font # Apply the Font to the Style
for i in range(7, len(good_bacteries.columns)):
    gb.write(0, i - 6, good_bacteries.columns[i], style = style)
    col = gb.col(i - 6)
    col.width = 256 * 25
gb.col(0).width = 256 * 25
for i in range(len(good_list)):
    gb.write(i + 1, 0, good_list[i], style = style)
#sum_for_cell = np.zeros((len(good_list), 6))
sum_for_cell = np.zeros((len(good_list), 7))
for index, row in good_bacteries.iterrows():
    if (row[1] == "Bacteroidetes"):
        for i in range(7, 14):
        #for i in range(7, 13):
            sum_for_cell[0][i - 7] += row[i]
    for j in range(len(good_list)):
        if (row[4] == good_list[j]):
            for i in range(7, 14):
            #for i in range(7, 13):
                sum_for_cell[j][i - 7] += row[i]
    if (type(row[4]) != float and row[4].find(search_clostrid) != -1):
        for i in range(7, 14):
            #for i in range(7, 13):
            sum_for_cell[5][i - 7] += row[i]
    if (row[3] == "Thermoanaerobacterales"):
        for i in range(7, 14):
            #for i in range(7, 13):
            sum_for_cell[6][i - 7] += row[i]
    if (row[3] == "Bifidobacteriales"):
        for i in range(7, 14):
            #for i in range(7, 13):
            sum_for_cell[10][i - 7] += row[i]
    if (row[3] == "Lactobacillales"):
        for i in range(7, 14):
            #for i in range(7, 13):
            sum_for_cell[11][i - 7] += row[i]
    if (row[3] == "Actinomycetales"):
        for i in range(7, 14):
            #for i in range(7, 13):
            sum_for_cell[12][i - 7] += row[i]
    if ((type(row[4]) != float and row[4].find(search_bacil) != -1)
       or (type(row[4]) != float and row[4].find(search_paen) != -1)):
        for i in range(7, 14):
            #for i in range(7, 13):
            sum_for_cell[9][i - 7] += row[i]


for i in range(sum_for_cell.shape[1]):
    sum_for_cell[7][i] += sum(sum_for_cell[:7,i])

for i in range(sum_for_cell.shape[0]):
    for j in range(sum_for_cell.shape[1]):
        gb.write(i + 1, j + 1, sum_for_cell[i][j], style = style)
    

"""
j = 1
for index, row in good_bacteries.iterrows():
    for i in range(len(row)):
        if (type(row[i]) == float and math.isnan(row[i])):
            gb.write(j, i, "", style = style)
        else:
            gb.write(j, i, row[i], style = style)
        if (j == 2 and i == 6):
            print(row[i] == float("NaN"))
    j += 1
"""
workbook.save('/Users/maksimveretennikov/Desktop/Coding/table_for_nicholas_2.xls')


```
